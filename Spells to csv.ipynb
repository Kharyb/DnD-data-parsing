```python
#import libraries
import json
import pandas as pd
```

```python
output_name = "converted_spells-xge.csv"
#open JSON
with open("spells-xge.json", "r") as read_json:
    data = json.load(read_json)
spells = data["spell"]
```

```python
#extract all spell names from JSON
spells_list = []
for s in spells:
    spells_list.append(s["name"])
```

```python
#create empty pandas dataframe to add columns to:
dfObj = pd.DataFrame(columns = ['Name' , 'Level', 'School' , 'Casting Time', 'Range', 'Components', 'Duration', 'Description','C', 'R', 'V', 'S', 'M', 'Metatitle', 'Metadescription'])
```

```python
#create levels dictionary
levels_dict = {"0" : "Cantrip",
               "1" : "1st Level",
               "2": "2nd Level",
               "3": "3rd Level",
               "4": "4th Level",
               "5": "5th Level",
               "6": "6th Level",
               "7": "7th Level",
               "8": "8th Level",
               "9": "9th Level"}
```

```python
schools_dict = {"V" : "Evocation",
               "T": "Transmutation",
               "N": "Necromancy",
               "I": "Illusion",
               "E": "Enchantment",
               "D": "Divination",
               "C": "Conjuration",
               "A": "Abjuration",
               "P": "Psionics"}
```

```python
#create loop over names that creates row in dataframe
for i in range(len(spells_list)):
    #generate file name
    #filename = str(spells_list[i]) + ".html"
    #generate file and open on write mode
    #Html_file= open(filename,"w")
    
    #find spell with this name in JSON and extract variables
    name = str(spells_list[i])
    
    level_num = str(spells[i]["level"])
    level = levels_dict[level_num]
    
    school_letter = str(spells[i]["school"])
    school = schools_dict[school_letter]
    
    #description = str(spells[i]["entries"])
    description = ""
    for j in range (len(spells[i]["entries"])):        
        try:
            current_entry_name = str(spells[i]["entries"][j]['name'])
            current_entry_text = str(spells[i]["entries"][j]['entries'][0])
            description += "\n"+ current_entry_name + ": "+ current_entry_text
        except:
            description += str((spells[i]["entries"][j]))
    description = description.replace("{@spell ", "")
    description = description.replace("}", "")
    description = description.replace("{@damage ", "")
    description = description.replace("{@item ", "")
    description = description.replace("{@filter ", "")
    description = description.replace("{@dice ", "")
    description = description.replace("{@condition ", "")
    description = description.replace("{@creature ", "")
    description = description.replace("{@hit ", "")
    description = description.replace("{@sense ", "")
    description = description.replace("{@action ", "")
    description = description.replace("{@skill ", "")
    
    cast_time = str(spells[i]["time"][0]['number']) + " "+ str(spells[i]["time"][0]['unit'])
    
    
    try:
        ritual = str(spells[i]["meta"]['ritual'])
        R="True"
        school = school + " (Ritual)"
    except:
        ritual = ""
        R = "False"
    
    try:
        range_ =str(spells[i]["range"]["distance"]["amount"]) + " "+ str(spells[i]["range"]["distance"]["type"])
    except:
        try:
            range_ = str(spells[i]["range"]["distance"]["type"])
        except:
            range_ = "Not found"
    
    components = ""
    #initiate false values
    V = "False"
    S = "False"
    M = "False"
    C = "False"
    
    try:
        if str(spells[i]["components"]['v']) == "True":
            components+= "V "
            V = "True"
    except:
        components
    try:
        if str(spells[i]["components"]['s']) == "True":
            components+= "S "
            S = "True"
    except:
        components
    try:
        components += "M ("+str(spells[i]["components"]['m']['text']) + ")"
        M = "True"
    except:
        try:
            components += "M ("+str(spells[i]["components"]['m']) + ")"
            M = "True"
        except:
            components += ""

    duration_type= str(spells[i]["duration"][0]["type"])
    if duration_type == "instant":
        duration = "Instantaneous"
    elif duration_type == "timed":
        duration = ""
        try:
            if (spells[i]["duration"][0]["concentration"]) == True:
                duration = "Concentration, "
                C = "True"
        except:
            duration
            
        if int(spells[i]["duration"][0]["duration"]["amount"]) >1:
            duration += str(spells[i]["duration"][0]["duration"]["amount"]) + " "+ str(spells[i]["duration"][0]["duration"]["type"])+"s"
        else: 
            duration += str(spells[i]["duration"][0]["duration"]["amount"]) + " "+ str(spells[i]["duration"][0]["duration"]["type"])
    
    metatitle = name + " - Spells | DnD Wiki"
    
    if level == "Cantrip": 
        if school[0] in ["A", "E", "I"]:
            metadescription = name + " is an " + school + " cantrip in Dungeons and Dragons 5th Edition (D&D 5E)."
        else:
            metadescription = name + " is a " + school + " cantrip in Dungeons and Dragons 5th Edition (D&D 5E)."
    else:
         metadescription = name + " is a "+ level+ " "+school+ " spell in Dungeons and Dragons 5th Edition (D&D 5E)."
    #create empty pandas dataframe to add columns to:
    #dfObj = pd.DataFrame(columns = ['Name' , 'Level', 'School' , 'Casting Time', 'Range', 'Components', 'Duration', 'Description'])
    dfObj = dfObj.append({'Name' : name , 
                          'Level' : level, 
                          'School' :school, 
                          'Casting Time' : cast_time, 
                          'Range' : range_, 
                          'Components' : components, 
                          'Duration' : duration, 
                          'Description': description,
                          'C' : C, 
                          'R' : R,
                          'V' : V, 
                          'S': S, 
                          'M': M,
                         'Metatitle' : metatitle,
                         'Metadescription' : metadescription}, ignore_index=True)
    #print(name + level + school + description + cast_time + range_+ components + duration)
```
