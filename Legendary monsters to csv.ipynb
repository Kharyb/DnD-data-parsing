```python
#import libraries
import json
import pandas as pd
```

```python
#create empty pandas dataframe to add columns to:
dfObj = pd.DataFrame(columns = ['Name' ,
                                'Lair Actions',
                                'Regional Effects'])
```

```python
#create size dictionary
size_dict = {"T" : "Tiny",
                  "S" : "Small",
                  "M": "Medium",
                  "L": "Large",
                  "H": "Huge",
                  "G": "Gargantuan"}
```

```python
#open JSON
with open("legendarygroups.json", "r") as read_json:
    data = json.load(read_json)
monsters = data["legendaryGroup"]
monsters
```

```python
#extract monster names
monsters_list = []
for m in monsters:
    monsters_list.append(m["name"])
```

```python

#create loop over names that creates row in dataframe
for i in range(len(monsters_list)):    
    #find monster with this name in JSON and extract variables
    name = str(monsters_list[i])
    
    #lair_actions
    try:
        #lair_actions_intro = str(monsters[i]['lairActions'][0])
        lair_actions = ""
        for j in range(len(monsters[i]['lairActions'])):
            if "items" in str(monsters[i]['lairActions'][j]):
                lair_actions_list = monsters[i]['lairActions'][j]["items"]
                for k in range(len(lair_actions_list)):
                    lair_actions += "\n"+ lair_actions_list[k]
                    
                #lair_actions += "\n"+ lair_actions_list

            else:
                lair_actions_item = str(monsters[i]['lairActions'][j])
                if j > 1:
                    lair_actions += "\n"+ lair_actions_item
                else:
                    lair_actions += lair_actions_item
                #return str(monsters[i]['lairActions'][j])
    except:
        lair_actions = "None"
    
    lair_actions = lair_actions.replace("{@condition ", "")
    lair_actions = lair_actions.replace("{@spell ", "")
    lair_actions = lair_actions.replace("{@dice", "")
    lair_actions = lair_actions.replace("}", "")
        
    #regional effects 
    try:
        #regional_effects_intro = str(monsters[i]['regionalEffects'][0])
        regional_effects = ""
        for j in range(len(monsters[i]['regionalEffects'])):
            if "items" in str(monsters[i]['regionalEffects'][j]):
                regional_effects_list = monsters[i]['regionalEffects'][j]["items"]
                for k in range(len(regional_effects_list)):
                    regional_effects += "\n"+ regional_effects_list[k]
                    
                #lair_actions += "\n"+ lair_actions_list

            else:
                regional_effects_item = str(monsters[i]['regionalEffects'][j])
                if j > 1:
                    regional_effects += "\n"+ regional_effects_item
                else:
                    regional_effects += regional_effects_item
                #return str(monsters[i]['lairActions'][j])
    except:
        regional_effects = "None"    
        
    regional_effects = regional_effects.replace("{@condition ", "")
    regional_effects = regional_effects.replace("{@spell ", "")
    regional_effects = regional_effects.replace("{@dice", "")
    regional_effects = regional_effects.replace("}", "")    
    #-----------------------------------------

    dfObj = dfObj.append({'Name': name ,
                            'Lair Actions': lair_actions,
                            'Regional Effects': regional_effects}, ignore_index=True)

    
```
