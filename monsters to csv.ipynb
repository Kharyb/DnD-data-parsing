```python
#import libraries
import json
import pandas as pd
```

```python
#open JSON
with open("bestiary-mm.json", "r") as read_json:
    data = json.load(read_json)
monsters = data["monster"]
```

```python
#extract all spell names from JSON
monsters_list = []
for m in monsters:
    monsters_list.append(m["name"])
```

```python
#create empty pandas dataframe to add columns to:
dfObj = pd.DataFrame(columns = ['Name' ,
                                'Type',
                                'Alignment', 
                                'Armor Class', 
                                'HP', 
                                'Speed', 
                                'Str', 
                                'Dex',
                                'Con', 
                                'Int', 
                                'Wis', 
                                'Cha', 
                                'Saving Throws',
                                'Skills',
                                'Damage Vulnerabilities',
                                'Damage Resistances',
                                'Damage Immunities',
                                'Condition Immunities',
                                'Senses',
                                'Languages',
                                'Challenge',
                                'Traits',
                                'Actions',
                                'Legendary Actions',
                                'Intro',
                                'MetaTitle',
                                'MetaDescription'])
```

```python
#create levels dictionary
alignment_dict = {"L" : "Lawful",
                  "N" : "Neutral",
                  "C": "Chaotic",
                  "G": "Good",
                  "E": "Evil",
                  "A": "Any Alignment",
                  "U" : "Unaligned"}
```

```python
#create loop over names that creates row in dataframe
for i in range(len(monsters_list)):    
    #find monster with this name in JSON and extract variables
    name = str(monsters_list[i])
    try:
        type_ = str(monsters[i]["type"]["type"])
    except:
        type_ = str(monsters[i]["type"])
    
    try:
        alignment0 = alignment_dict[str(monsters[i]["alignment"][0])]
        alignment1 = alignment_dict[str(monsters[i]["alignment"][1])]
        alignment = alignment0 + " "+ alignment1
        if alignment == "Neutral Neutral":
            alignment = "True Neutral"
    except:
        try:
            alignment = alignment_dict[str(monsters[i]["alignment"][0])]
        except:
            alignment = "error"
                                
    armor_class = str(monsters[i]["ac"][0])
    if len(armor_class) > 3:
        try:
            armor_class = str(monsters[i]["ac"][0]["ac"]) + " (" + str(monsters[i]["ac"][0]["from"][0])+ ")"
        except:
            armor_class = "error"
    armor_class = armor_class.replace("{@item", "")
    armor_class = armor_class.replace("|phb}", "")
    
    hp_average = str(monsters[i]["hp"]["average"])
    hp_formula = str(monsters[i]["hp"]["formula"])
    hp = hp_average + " ("+hp_formula + ")"
    
    try:
        walk_speed = str(monsters[i]["speed"]["walk"]) + "ft."
        fly_speed = ", fly " + str(monsters[i]["speed"]["fly"]) + "ft."
        swim_speed = ", swim " + str(monsters[i]["speed"]["swim"]) + "ft."

        speed = walk_speed + fly_speed + swim_speed
    except:
        try:
            walk_speed = str(monsters[i]["speed"]["walk"]) + "ft."
            fly_speed = ", fly "+ str(monsters[i]["speed"]["fly"]) + "ft."
            speed = walk_speed + fly_speed
        except:
            try:
                walk_speed = str(monsters[i]["speed"]["walk"]) + "ft."
                swim_speed = ", swim " + str(monsters[i]["speed"]["swim"]) + "ft."
                speed = walk_speed + swim_speed
            except:
                try:
                    walk_speed = str(monsters[i]["speed"]["walk"]) + "ft."
                    speed = walk_speed
                except:
                    speed = "error"
    Str = str(monsters[i]["str"])
    Dex = str(monsters[i]["dex"])
    Con = str(monsters[i]["con"])
    Int = str(monsters[i]["int"])
    Wis = str(monsters[i]["wis"])
    Cha = str(monsters[i]["cha"])
    
    try:
        skills = str(monsters[i]["skill"]).replace("{", "")
        skills = skills.replace("}", "")
        skills = skills.replace("'", "")
    except:
        skills = "None"
    
    try:
        senses = str(monsters[i]["senses"])
        senses = senses.replace("'", "")
        senses = senses.replace("[", "")
        senses = senses.replace("]", "")
    except:
        senses = "error"
    
    try:
        languages = str(monsters[i]["languages"])
        languages = languages.replace("'", "")
        languages = languages.replace("[", "")
        languages = languages.replace("]", "")
    except:
        languages = "error"
        
    challenge = str(monsters[i]["cr"])
    
    try:
        traits = ""
        for j in range (len(monsters[i]["trait"])):        
            try:
                current_entry_name = str(monsters[i]["trait"][j]['name'])
                current_entry_text = str(monsters[i]["trait"][j]['entries'][0])
                traits += "\n"+ current_entry_name + ": "+ current_entry_text
                
                
                
            except:
                traits += str(monsters[i]["trait"][j])
        traits = traits.replace("{@spell ", "")
        traits = traits.replace("}", "")
        traits = traits.replace("{@damage ", "")
        traits = traits.replace("{@item ", "")
        traits = traits.replace("{@filter ", "")
        traits = traits.replace("{@dice ", "")
        traits = traits.replace("{@condition ", "")
        traits = traits.replace("{@creature ", "")
        traits = traits.replace("{@hit ", " Hit: ")
        traits = traits.replace("{@sense ", "")
        traits = traits.replace("{@action ", "")
        traits = traits.replace("{@skill ", "")
    except:    
        traits = "error"
    
    try:
        actions = ""
        for k in range (len(monsters[i]["action"])):        
            try:
                current_action_name = str(monsters[i]["action"][k]['name'])
                current_action_text = str(monsters[i]["action"][k]['entries'][0])
                actions += "\n"+ current_action_name + ": "+ current_action_text
                
                
                
            except:
                actions += str(monsters[i]["action"][k])
        actions = actions.replace("{@spell ", "")
        actions = actions.replace("{@damage ", "")
        actions = actions.replace("{@item ", "")
        actions = actions.replace("{@filter ", "")
        actions = actions.replace("{@dice ", "")
        actions = actions.replace("{@condition ", "")
        actions = actions.replace("{@creature ", "")
        actions = actions.replace("{@hit ", "+")
        actions = actions.replace("{@h", "Hit: ")
        actions = actions.replace("{@sense ", "")
        actions = actions.replace("{@action ", "")
        actions = actions.replace("{@skill ", "")
        actions = actions.replace("{@atk mw} ", "Melee Weapon Attack: ")
        actions = actions.replace("{@atk rw} ", "Ranged Weapon Attack: ")
        actions = actions.replace("{@atk mw,rw} ", "Melee or Ranged Weapon Attack: +")
        actions = actions.replace("{@dc ", "DC: ")
        actions = actions.replace("{@recharge", " - Recharge ")
        actions = actions.replace("}", "")
        
    except:    
        actions = "error"
        
    
    try:
        legendary_actions = ""
        for l in range (len(monsters[i]["legendary"])):        
            try:
                current_laction_name = str(monsters[i]["action"][l]['name'])
                current_laction_text = str(monsters[i]["action"][l]['entries'][0])
                legendary_actions += "\n"+ current_laction_name + ": "+ current_laction_text          
                
            except:
                legendary_actions += str(monsters[i]["action"][l])
        legendary_actions = legendary_actions.replace("{@spell ", "")
        legendary_actions = legendary_actions.replace("{@damage ", "")
        legendary_actions = legendary_actions.replace("{@item ", "")
        legendary_actions = legendary_actions.replace("{@filter ", "")
        legendary_actions = legendary_actions.replace("{@dice ", "")
        legendary_actions = legendary_actions.replace("{@condition ", "")
        legendary_actions = legendary_actions.replace("{@creature ", "")
        legendary_actions = legendary_actions.replace("{@hit ", "+")
        legendary_actions = legendary_actions.replace("{@h", "Hit: ")
        legendary_actions = legendary_actions.replace("{@sense ", "")
        legendary_actions = legendary_actions.replace("{@action ", "")
        legendary_actions = legendary_actions.replace("{@skill ", "")
        legendary_actions = legendary_actions.replace("{@atk mw} ", "Melee Weapon Attack: ")
        legendary_actions = legendary_actions.replace("{@atk rw} ", "Ranged Weapon Attack: ")
        legendary_actions = legendary_actions.replace("{@atk mw,rw} ", "Melee or Ranged Weapon Attack: +")
        legendary_actions = legendary_actions.replace("{@dc ", "DC: ")
        legendary_actions = legendary_actions.replace("{@recharge", " - Recharge ")
        legendary_actions = legendary_actions.replace("}", "")
        
    except:    
        legendary_actions = "None"

        
    try:
        save_throws = str(monsters[i]["save"])
        save_throws = save_throws.replace("str", "Str")
        save_throws = save_throws.replace("dex", "Dex")
        save_throws = save_throws.replace("con", "Con")
        save_throws = save_throws.replace("int", "Int")
        save_throws = save_throws.replace("wis", "Wis")
        save_throws = save_throws.replace("cha", "Cha")
        save_throws = save_throws.replace("'", "")
        save_throws = save_throws.replace("{", "")
        save_throws = save_throws.replace("}", "")
        
    except:
        save_throws = "None"
        
    try:
        damage_immun = str(monsters[i]["immune"])
        damage_immun = damage_immun.replace("'", "")
        damage_immun = damage_immun.replace("]", "")
        damage_immun = damage_immun.replace("[", "")
        
    except:
        damage_immun = "None"
        
    
    try:
        condition_immun = str(monsters[i]["conditionImmune"])
        condition_immun = condition_immun.replace("'", "")
        condition_immun = condition_immun.replace("]", "")
        condition_immun = condition_immun.replace("[", "")
        
    except:
        condition_immun = "None"
        
    try:
        damage_resist = str(monsters[i]["resist"])
        damage_resist = damage_resist.replace("'", "")
        damage_resist = damage_resist.replace("]", "")
        damage_resist = damage_resist.replace("[", "")
        
    except:
        damage_resist = "None"
        
    try:
        damage_vulns = str(monsters[i]["vulnerable"])
        damage_vulns = damage_vulns.replace("'", "")
        damage_vulns = damage_vulns.replace("]", "")
        damage_vulns = damage_vulns.replace("[", "")
        
    except:
        damage_vulns = "None"
        
    
    #---------------------------------------------------------
    
    intro = "not in this JSON"
                                                             
    metatitle = name + " - Monsters | DnD Wiki"
    metadescription = name + " is a CR " + challenge + ", " +alignment+ ", "+ type_ + ", in Dungeons and Dragons 5th Edition (D&D 5e)"
    
    #create empty pandas dataframe to add columns to:
    #dfObj = pd.DataFrame(columns = ['Name' , 'Level', 'School' , 'Casting Time', 'Range', 'Components', 'Duration', 'Description'])
    dfObj = dfObj.append({'Name': name ,
                            'Type': type_,
                            'Alignment': alignment, 
                            'Armor Class': armor_class, 
                            'HP': hp,  
                            'Speed' : speed,
                            'Str' : Str , 
                            'Dex' : Dex,
                            'Con' : Con, 
                            'Int' : Int, 
                            'Wis' : Wis, 
                            'Cha' : Cha, 
                            'Saving Throws' : save_throws,
                            'Skills' : skills,
                            'Damage Vulnerabilities' : damage_vulns,
                            'Damage Resistances' : damage_resist,
                            'Damage Immunities' : damage_immun,
                            'Condition Immunities' : condition_immun,
                            'Senses' : senses,
                            'Languages' : languages,
                            'Challenge' : challenge,
                            'Traits' : traits,
                            'Actions' : actions,
                            'Legendary Actions' : legendary_actions,
                            'Intro' : intro,
                            'MetaTitle' : metatitle,
                            'MetaDescription' : metadescription}, ignore_index=True)
    #print(name + level + school + description + cast_time + range_+ components + duration)
```
