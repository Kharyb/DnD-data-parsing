```python
#import libraries
import json
import pandas as pd
```

```python
#open JSON
with open("fluff-bestiary-mm.json", "r") as read_json:
    data = json.load(read_json)
monsters = data["monsterFluff"]
```

```python
#extract all spell names from JSON
monsters_list = []
for m in monsters:
    monsters_list.append(m["name"])
```

```python
#create empty pandas dataframe to add columns to:
dfObj = pd.DataFrame(columns = ['Name' ,
                                'Info'])
```

```python
#create levels dictionary
alignment_dict = {"L" : "Lawful",
                  "N" : "Neutral",
                  "C": "Chaotic",
                  "G": "Good",
                  "E": "Evil",
                  "A": "Any Alignment"}
```

```python
#create loop over names that creates row in dataframe
for i in range(len(monsters_list)):    
    #find monster with this name in JSON and extract variables
    name = str(monsters_list[i])
    '''
    try:
        p1 = str(monsters[i]["entries"][0]['entries'][0]['entries'])
        
    except:
        p1 = "no entries"
    '''
    try:
        info = ""
        for j in range (len(monsters[i]["entries"][0]['entries'][0]['entries'])):        
            try:
                current_entry_name = str(monsters[i]["entries"][0]['entries'][0]['entries'][j]['name'])
                current_entry_text = str(monsters[i]["entries"][0]['entries'][0]['entries'][j]['entries'][0])
                info += "\n"+ current_entry_name + ": "+ current_entry_text
                
                
                
            except:
                info += str(monsters[i]["entries"][0]['entries'][0]['entries'][j])
    except:
        info = "No entries"
            
    
    #---------------------------------------------------------
    
    #create empty pandas dataframe to add columns to:
    #dfObj = pd.DataFrame(columns = ['Name' , 'Level', 'School' , 'Casting Time', 'Range', 'Components', 'Duration', 'Description'])
    dfObj = dfObj.append({'Name': name ,
                            'Info': info}, ignore_index=True)
    #print(name + level + school + description + cast_time + range_+ components + duration)
```
